```python
import yfinance as yf
import pandas as pd

# Fetch last 60 days, 5-min interval
data = yf.download("AAPL", period="60d", interval="5m")

print(data.head())
print(data.tail())
```

```python
#save with Datetime as a normal column (no index in CSV)
df = data.reset_index() 
df.to_csv("aapl_5min_60d_raw.csv", index=False)
```

```python
import pandas_ta as ta

data_ta = data.copy()
data_ta = data_ta.ta.add_all_ta_features(open="Open", high="High", low="Low", close="Close", volume="Volume")

# move datetime index to column and save
data_ta.reset_index(inplace=True)
data_ta.to_csv("aapl_5min_60d_ta.csv", index=False)
```
